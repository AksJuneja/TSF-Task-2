```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Reading data using link

```python
data = pd.read_csv("http://bit.ly/w-data")
data
```

## Plotting the distribution 

```python
data.plot(x='Hours', y='Scores', style='x')
plt.title('The distribution curve with Hours vs percentage')
plt.xlabel('Hours')
plt.ylabel('Percentage Score')
plt.show()
```

## Preparing Data
Lets divide data into attributes and labels i.e, inputs and outputs

```python
X= data.iloc[:, :-1].values
y= data.iloc[:, 1].values
```

## Splitting Data into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Training  

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Plotting the regression line
line= regressor.coef_*X+regressor.intercept_
#Plotting the test data
plt.scatter(X, y)
plt.plot(X, line)
plt.show()
```

## Predictions

```python
#testing data- in hours
print(X_test)
#predicting the scores
y_pred = regressor.predict(X_test)
```

```python
#Comparing Actual vs Predicted 
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

## Predicting the score of the student with 9.25 hours studying

```python
hours =[[9.25]]
ques_pred = regressor.predict(hours)
print('No of Hours={}'.format(hours))
print('Predicted Score={}'.format(ques_pred[0]))
```

## Evaluating Model

```python
from sklearn import metrics
print('Mean Absolute Error=', metrics.mean_absolute_error(y_test, y_pred))
```
